```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
import math
```

```python
def sigmoid(x):
    return 1/ (1 + math.e ** -x)
def sigmoid_derivative(x):
    return (sigmoid(x) * (1-sigmoid(x)))
```

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
print(x)
print(y)
```

```python
epochs = 5000
input_size , hidden_size , output_size = 2 , 3, 1
learning_rate = 0.1
```

```python
W1 = np.random.uniform(size = (input_size,hidden_size))
W2 = np.random.uniform(size = (hidden_size,output_size))
```

```python
def doComputation(W1,W2,learning_rate):
    z1 = np.dot(x,W1)
    A1 = sigmoid(z1)
    #print(A1)
    z2 = np.dot(A1,W2)
    y_hat = sigmoid(z2)
    #print(y_hat)
    E = y_hat - y
    #print(E)
    learning_rate = learning_rate*E
    return E,learning_rate
```

```python
learning_rate = .1
for i in range(epochs):
    delta1 = E*learning_rate #1
    W2  -= np.dot(A1.T,delta1)  #2
    #print(delta1.shape)
    #print(W2.shape)
    delta2 = np.dot(delta1,W2.T) * sigmoid_derivative(A1) #3
    #print(delta2.shape)
    #print(x.T.shape)
    W1 -= np.dot(x.T,delta2)#4
    E,learning_rate = doComputation(W1,W2,learning_rate)
    print(E,end="\n\n")
```
