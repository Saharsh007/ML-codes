```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import copy
```

```python
df = pd.DataFrame({
    'x': [12, 20, 28, 18, 29, 33, 24, 45, 45, 52, 51, 52, 55, 53, 55, 61, 64, 69, 72],
    'y': [39, 36, 30, 52, 54, 46, 55, 59, 63, 70, 66, 63, 58, 23, 14, 8, 19, 7, 24]
})
```

```python
df.head()
```

```python
k = 3
# centroids[i] = [x, y]
centroids = {
    i+1: [np.random.randint(0, 80), np.random.randint(0, 80)]
    for i in range(k)
}
```

```python
print(centroids)
print(type(df['x']))
```

```python
fig= plt.figure(figsize=(5,5))
plt.scatter(df['x'],df['y'],color = 'k')
colmap = {1:'r' , 2:'g',3:'b'}
for i in centroids.keys():
    plt.scatter(centroids[i][0],centroids[i][1],color = colmap[i])
plt.xlim(0,80)
plt.ylim(0.80)
plt.show()
```

```python
def calculateDistance(df,centroids):
    for i in centroids.keys():
        df["distance_from_{}".format(i)] =(\
            np.sqrt( (df['x'] - centroids[i][0] )**2  + (df['y'] - centroids[i][1]) ** 2 ) )
        
    return df
```

* format is used to concaternate string in that specific pattern  
  https://www.geeksforgeeks.org/python-format-function/  

* this creates 3 new columns in df distance_from_1 ,2, 3 with respective distances for all the points  


```python
df = calculateDistance(df,centroids)
df.head()
```

* next is to find closest centroid for each point and assign the same color

<img src="1.png" align="left" alt="Drawing" style="width: 800px; height:400px"/>

```python
def assign_nearest(df,centroids):
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df["nearest"] = df.loc[: , centroid_distance_cols].idxmin(axis = 1)
    df["nearest"] = df["nearest"].apply(lambda x: int (x.lstrip("distance_from_")) )
    df["color"] = df["nearest"].map(lambda x : colmap[x])
    return df
```

* idxmin - Return index of first occurrence of minimum over requested axis. NA/null values are excluded.  
* df[nearest].map Used for substituting each value in a Series with another value, that may be derived from a function, a dict or a Series.



```python
df = assign_nearest(df,centroids)
df.head()
```

```python
def assignment_graph():
    fig = plt.figure(figsize=(5,5))
    plt.scatter(df['x'] , df['y'] , alpha=0.4 , color= df['color'])
    for i in  centroids.keys():
        plt.scatter(centroids[i][0] , centroids[i][1] , color= colmap[i])
    plt.show
    
assignment_graph()
```

##  Part2 - Updating location of centroids

```python
import copy
```

```python
old_centroids = copy.deepcopy(centroids)
```

 * df[df['nearest'] == 3]['x'] - how is this working ? -> (https://stackoverflow.com/questions/17071871/select-rows-from-a-dataframe-based-on-values-in-a-column-in-pandas/17071908)

```python
i = 1
def update_Values(centroids):
    for i in centroids.keys():
        centroids[i][0] = int( np.mean(df[df["nearest"] == i]['x']) )
        centroids[i][1] = int( np.mean(df[df["nearest"] == i]['y']) )
    return centroids
```

```python
centroids = update_Values(centroids)
print(centroids)
```

```python
def change_graph(old_centroids):
    fig = plt.figure(figsize=(5,5))
    ax = plt.axes()
    plt.xlim(0,80)
    plt.ylim(0,80)
    plt.scatter(df['x'] , df['y'] , alpha=0.4 , color= df['color'])
    for i in  centroids.keys():
        plt.scatter(centroids[i][0] , centroids[i][1] , color= colmap[i])


    for i in centroids.keys():
        old_x = old_centroids[i][0]
        old_y = old_centroids[i][1]
        dx = (centroids[i][0] - old_centroids[i][0]) * 0.75
        dy = (centroids[i][1] - old_centroids[i][1]) * 0.75
         #changes the corrdinate to some distance behind the new coordinate
        ax.arrow(old_x, old_y, dx, dy, head_width=2, head_length=3, fc=colmap[i], color=colmap[i])
    plt.show
change_graph(old_centroids)
```

## Part3 : repeat untill convergence

```python
while True:
    before_change = df['nearest'].copy(deep=True) #does not change if centroids changes
    old_centroids = copy.deepcopy(centroids)
    centroids = update_Values(centroids)
    change_graph(old_centroids)
    df = assign_nearest(df,centroids)
    assignment_graph()
    if before_change.equals(df["nearest"]):
        break
centroids
```

# Now we will use sklearn

```python
df = pd.DataFrame({
    'x': [12, 20, 28, 18, 29, 33, 24, 45, 45, 52, 51, 52, 55, 53, 55, 61, 64, 69, 72],
    'y': [39, 36, 30, 52, 54, 46, 55, 59, 63, 70, 66, 63, 58, 23, 14, 8, 19, 7, 24]
})

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
print(kmeans)
labels = kmeans.predict(df)
centroids = kmeans.cluster_centers_
centroids
```
