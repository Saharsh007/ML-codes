```python
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import movie_reviews
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
def convertIntoNaiveBayersFormat(text):
    final = dict([(word,True) for word in text if word not in stopwords.words('english')\
                 and word not in string.punctuation and word.isdigit() == False  ])
    return final

#type casting for dictionary instead of list
#without typecasting this would have happened
#[('Let', True), ('us', True), ('see', True), ('works', True)]
#why dictionary ?
#it'll won't count repeated words
```

```python
negReviews = []
for fileid in  movie_reviews.fileids('neg'):
    words = movie_reviews.words(fileid)
    negReviews.append((convertIntoNaiveBayersFormat(words),"negative"))
```

```python
negReviews[0]
```

```python
posReviews = []
for fileid in movie_reviews.fileids('pos'):
    word = movie_reviews.words(fileid)
    posReviews.append((convertIntoNaiveBayersFormat(word) , "positive") )
```

```python
posReviews[0]
```

```python
trainSet = negReviews[:750] + posReviews[:750]
testSet = negReviews[750:] + posReviews[750:]
```

* And we are all done now we can use Naive bayer algorithm in train test and check accuracy on testset

```python
classifier = NaiveBayesClassifier.train(trainSet)
```

* classifier contains the trained model on 1500 examples
* we can use it for prediction now

```python
print(classifier)
```

####  We will use nltk classifier and compare our model with it 

```python
accuracy  = nltk.classify.util.accuracy(classifier,testSet)
accuracy*100
```
